## Functions to Create the Dataset

```python
# Function to load the necessary Dataset. Returns a dataframe containing
# stock information
def stock_loader(s : str, duration : str):
    """
    Function to load the necessary stock data based on name and duration

    Parameters:
        s: String for the name of the dataset
        duration: Time duration of stock data required
    Returns:
        stock_df: Dataframe containing the stock price data
    """
    try:
        ticker = yf.Ticker(s)
        stock_df = ticker.history(period = duration)
    except:
        return "Stock name or duration given is invalid"
    ## Plotting stock
    stock_df["Close"].plot(title = "{}'s stock price".format(s))
    return stock_df
```

```python
# Function to generate time series data
def datasetgen(dataset, window_size: int):
    """
    Function to generate windowed time series dataset
    
    Parameters
        dataset: A numpy array containing the stock price across time
        window_size: An integer signifying the size of the context window

    Return:
        X: List containing the input stock price data. Length = length of window
        Y: List containing the prediction stock price data -- predicting length of window + 1
    """
    if not isinstance(dataset, np.ndarray):
        return "Dataset should be an numpy array of closing price vals"

    n = dataset.shape[0]
    X, Y = [], []
    # Window size determines the lookback window. If 20, we start from the 20th
    # index to ensure 20 previous values
    for i in range(window_size, n):
        # X will not be normalized as we will turn it into fuzzy values
        x = dataset[i - window_size : i]
        # Only Y will be normalized
        # y = (dataset[i] - np.min(dataset)) / (np.max(dataset) - np.min(dataset))
        y = dataset[i]
        X.append(x)
        Y.append(y)
    return X, Y
```

## Functions for Fuzzy

```python
# Defining the Gaussian Function
def gaussmf(x, params):
    """
    Function that provides the gaussian for a set of datapoints 
    based on parameters Mean and standard deviation

    Parameters:
        x: Numpy array containing input values
        params: List containing the Mean and Standard Deviation

    Returns:
        Gaussian function value of input x using the parameters
    """
    assert len(params) == 2, 'Gaussmf function must have 2 parameters.'
    if not isinstance(x, np.ndarray):
        return "Input should be a numpy array"
    sig, c = np.asarray(params)
    return np.exp(-pow((x - c), 2) / (2 * pow(sig, 2)))

# A 2-Gaussian Function
def gauss2mf(x, params):
    """
    Function that provides the gaussian for a set of datapoints 
    based on parameters Mean and standard deviation

    Parameters:
        x: Numpy array containing input values
        params: List containing the Mean and Standard Deviation

    Returns:
        y: Gaussian function value of input x using the parameters
    """
    assert len(params) == 4, 'Gauss2mf function must have 4 parameters.'
    if not isinstance(x, np.ndarray):
        return "Input should be a numpy array"
    sig1, c1, sig2, c2 = np.asarray(params)
    assert c1 <= c2, 'c1 <= c2 is required.'
    if np.isscalar(x):
        if x<=c1:
            y=gaussmf(x, [sig1, c1])
        else:
            y=gaussmf(x,[sig2, c2])
    else:
        y = np.ones(len(x))
        idx1 = (x <= c1)
        idx2 = (x > c2)
        y[idx1] = gaussmf(x[idx1], [sig1, c1])
        y[idx2] = gaussmf(x[idx2], [sig2, c2])
    return y
```

```python
# Code taken from Assignment 4. Helps in doing inference with fuzzy
class fuzzymf(object):
    def __init__(self, Type, Parameters):
        """
        Initializes the fuzzy membership function with the given parameters

        Parameters:
            Type: The type of membership function to use. 
            Parameters: The parameters of the given membership function
        """
        self.Type = Type
        self.Parameters = Parameters
    def __repr__(self):
            return 'fismf, '\
                ' Type: %s, '\
                ' Parameters: %s\n'\
                % (self.Type,self.Parameters)

def evalmf(mf, x):
    """
    A dynamic function, that retrieves the membership function based on the input

    Parameters:
        mf: fuzzymf class containing the type and parameters of the membership function
        x: Numpy array containing the input dataset
    Returns:
        Transformed dataset in the form of numpy array
    """
    possibles = globals().copy()
    possibles.update(locals())
    method = possibles.get(mf.Type)
    return method(x, mf.Parameters)
```

```python
# LVQ Model for Fuzzy training
def lvq_train(X, nlabels, learning_rate=0.01, epochs=40):
    """
    Function to run training for Learning Vector Quantization

    Parameters:
        X: The dataset numpy array
        nlabels: The number of labels we will classify the dataset into 
        learning_rate: Learning rate parameter for updating weights
        epochs: Number of dataset passes
    Returns:
        prototypes: Numpy array with the final weights
    """
    # Randomly initialize prototypes from the training data
    np.random.seed(0)  # For reproducibility
    # Random index pertaining to number of random weights needed
    prototypes = X[np.random.choice(X.shape[0], nlabels, replace=False)]
    #print("Starting Weights: ", prototypes)
    for epoch in range(epochs):
        for sample in X:
            # Calculate distances to prototype
            if len(X.shape) == 1:
              distances = np.linalg.norm(sample - prototypes)
            else:
              distances = np.linalg.norm(sample - prototypes, axis=1)
            winner_idx = np.argmin(distances)  # Find the index of the closest prototype

            # Update the prototype based on classification
            # Here you can modify the logic based on class labels if necessary
            prototypes[winner_idx] += learning_rate * (sample - prototypes[winner_idx])
    #print("Final Weights: ", prototypes)
    return prototypes
```

```python
# Creating Membership functions with LVQ
def LVQ_learnmem(x, params):
    """
    A function to create membership functions using Learning Vector Quantization

    Parameters:
        x: The dataset numpy array
        params: Parameters required for creating the membership function

    Returns:
        fis: List with the membership functions
    """
    assert len(params) == 2, 'spam_learnmem function must have 2 parameters.'
    nlabels, tail = np.asarray(params)
    nlabels=int(nlabels)
    tail=bool(tail)
    # Obtaining optimized centroids
    prototypes = lvq_train(x, nlabels)
    # If dataset dimension is 1 -- when does this happen ? When we use only one stock price or differencing  
    if len(x.shape) == 1:
      ninputs = 1
    else:
      ninputs = x.shape[1]
    fis = []
    # For each dim, we will curate membership functions
    for i in range(ninputs):
      if ninputs == 1:
        centers = prototypes
      else:
        centers = prototypes[:, i]
      # Creating the centroids
      minx, maxx = np.min(centers), np.max(centers)
      centroids = np.linspace(minx, maxx, int(nlabels))
      sig = (centroids[1] - centroids[0]) / (2 * np.sqrt(2 * np.log(2)))
      mf=[]
      # Creating each membership function in the chosen dimension
      for j in range(nlabels):
          if tail:
            if j == 0:
              # Leftmost
              mf.append(fuzzymf(Type = "gauss2mf", Parameters = [sig, -np.inf, sig, centroids[j]]))
            elif j == nlabels - 1:
              # Rightmost
              mf.append(fuzzymf(Type = "gauss2mf", Parameters = [sig, centroids[j], sig, np.inf]))
            else:
              # Middle
             mf.append(fuzzymf(Type = 'gauss2mf', Parameters = [sig, centroids[j], sig, centroids[j]]))
          else:
             mf.append(fuzzymf(Type = 'gauss2mf', Parameters = [sig, centroids[j], sig, centroids[j]]))
      fis.append(mf)
    return fis
```

```python
# Fuzzify your input
# The inputs are ur array, the membership function and the number of membership
# functions use
def fuzzification(x, mf, membership: int):
    """
    Function to fuzzify your dataset

    Parameters:
        x: The dataset numpy array
        mf: List of membership functions
        membership: Number of membership functions
    Return:
        output_array: Fuzzified version of the dataset
    """
    assert len(x.shape) == 2
    dims = x.shape[1]
    output_array = np.zeros((membership, dims))
    # Fuzzifying each dim
    for i in range(dims):
        for j in range(membership):
            val = evalmf(mf[0][j], x[:, i])
            output_array[j, i] = val
    return output_array
```

```python
# Data loader for the Attention-LSTM
class CustomFuzzyDataset(Dataset):
    def __init__(self, x_list, y_list):
        """
        Initializes the dataset with input and output lists.

        Parameters:
        - x_list: List of input sequences (each sequence should have length 5).
        - y_list: List of corresponding output values.
        """
        self.x_data = x_list
        self.y_data = y_list

    def __len__(self):
        """Returns the total number of samples in the dataset."""
        return len(self.x_data)

    def __getitem__(self, idx):
        """
        Fetches the input-output pair at the specified index.

        Parameters:
        - idx: Index of the sample to fetch.

        Returns:
        A tuple containing the input tensor and the corresponding output tensor.
        """
        x = torch.tensor(self.x_data[idx], dtype = torch.float32).permute(1, 0)
        y = torch.tensor(self.y_data[idx], dtype = torch.float32)
        return x, y
```

**Terrific LSTM blog: https://colah.github.io/posts/2015-08-Understanding-LSTMs/?source=post_page-----37e2f46f1714--------------------------------**

**LSTM Model logic:**


$$Input Gate: i_t = Sigmoid(W_{ii} * x_t + b_{ii} + W_{ih} * h_{t-1} + b_{ih})$$
$$Forget Gate: f_t = Sigmoid(W_{if} * x_t + b_{if} + W_{hf} * h_{t-1} + b_{fh})$$
$$Cell Gate: c_t = Tanh(W_{ic} * x_t + b_{ic} + W_{hc} * h_{t-1} + b_{hc})$$
$$Output Gate: o_t = Sigmoid(W_{io} * x_t + b_{io} + W{ho} * h_{t-1} + b_{ho})$$

1. The Forget gate determines how much of the previous cell state $c_{t - 1}$ needs to be forgotten. 
So, the first step in an LSTM is: $$c_{t} = f_t * C{t-1}$$

2. Next, we need to update the values for the cell at $c_t$. For this, we take the output of the Input gate and multiply it by the output of the Cell gate: $i_t * c_t$. We use tanh to create new candidate values for the cell state, allowing for both positive and negative contributions. The input gate (using sigmoid) then determines how much of these new values should be added to the cell state. This product is then added to the previous cell state to obtain the new cell state: $$c_t = f_t * c_{t-1} + i_t * c_t$$

3. Next, we determine the hidden state of the cell. For this, we use the output of the Output gate $o_t$ and the current cell state $c_t$: $$h_t = o_t * Tanh(c_t)$$

4. In a PyTorch LSTM model object, the outputs are the following:

    a. **Output** of shape (Batch Size, Sequence Length, D * $H_{out}$), where D is 2 if bi-directional and 1 if not; H_out is the hidden size dimension.

   
    b. **$h_n$**, which is the hidden state of the final layer, of shape (D$*$num_layers, batch_size, $H_{out}$).

   
    c. **$c_n$**, which is the final cell state, of shape (D$*$num_layers, Batch Size, $H_{cell}$.

5. For both $h_n$ and $c_n$, the reason why we have the number of layers in the dimension is because all the hidden and cell states across the layers are stacked. 
   
   
**Bahdanau (Additive) Attention Model logic:**

$$Output = V_a * Activation(W_a * input + U_a * input)$$

1. The attention is applied on the LSTM output, which is of shape (Batch Size, Sequence Length, D * $H_{out}$, where $H_{out}$ is the hidden dimension.
   
2. The goal in using attention is for addressing the long-term dependency problem: For long sequence lengths, since the LSTM performs its operations per sequence element, there is no guarantee information from the first sequence element will be available to interact with the information from the last sequence element.

3. The softmax activation is applied across the sequence dimension, giving us weights for each sequence position. This helps us "attend" to each sequence position.

```python
# Define the attention module -- AI generated
class Attention(nn.Module):
    def __init__(self, hidden_size):
        """
        Function to initialize an additive Attention layer

        Parameters:
            hidden_size: Dimension of the hidden size
        """
        super(Attention, self).__init__()
        self.hidden_size = hidden_size
        # Feed forward layers to create the weight vectors
        self.Wa = nn.Linear(hidden_size, hidden_size)
        self.Ua = nn.Linear(hidden_size, hidden_size)
        self.Va = nn.Linear(hidden_size, 1)
        # Initializing weights
        self.initialize_weights()
    def forward(self, lstm_out):
        """
        Function to run forward pass with attention layer

        Parameters:
            lstm_out: Output from the lstm
        Returns:
            context_vector: Shape = (batch_size, hidden_size)
            attention_weights: Shape = (batch_size, seq_length, 1)
        """
        # lstm_out shape: (batch_size, seq_length, hidden_size)
        # Calculate attention scores
        scores = self.Va(torch.tanh(self.Wa(lstm_out) + self.Ua(lstm_out)))
        attention_weights = torch.softmax(scores, dim=1)  # Shape: (batch_size, seq_length, 1)

        # Weighted sum of the LSTM outputs
        context_vector = torch.bmm(attention_weights.permute(0, 2, 1), lstm_out)  # Shape: (batch_size, 1, hidden_size)
        context_vector = context_vector.squeeze(1)  # Shape: (batch_size, hidden_size)

        return context_vector, attention_weights

    def initialize_weights(self):
        """
        Function to initialize the weights of the Attention Layers
        """
        for name, param in self.named_parameters():
            if 'weight' in name:
                nn.init.kaiming_uniform_(param.data)  # Kaiming uniform initialization
            elif 'bias' in name:
                nn.init.zeros_(param.data)  # Initialize biases to zero


# Define the LSTM Model - AI generated
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        """
        Function to initialize the LSTM model
        Parameters
            input_size: The dimension of the input data
            hiiden_size: The dimension of the hidden layer
            num_layers: The number of hidden layers
        """
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.attention = Attention(hidden_size)
        self.fc = nn.Linear(hidden_size, 1)  # Output layer for 1D output
        # Initializing weight
        self.initialize_weights()
    def forward(self, x):
        """
        Function to run forward pass
        Parameters
            x: Input
        Return 
            output: Tensor corresponding to the output
        """
        # x shape: (batch_size, seq_length, input_size)
        lstm_out, (h_n, c_n) = self.lstm(x)  # Get the LSTM output
        context_vector, attention_weights = self.attention(lstm_out)  # Apply attention mechanism
        output = self.fc(context_vector)  # Shape: (batch_size, 1)
        return output

    def initialize_weights(self):
        """
        Function to initialize weights of the LSTM layers
        """
        # Initialize weights for LSTM layers
        for name, param in self.named_parameters():
            if 'weight' in name:
                nn.init.kaiming_uniform_(param.data)  # Kaiming uniform initialization
            elif 'bias' in name:
                nn.init.zeros_(param.data)  # Initialize biases to zero

```

## Cross Validation Function

```python
def train_and_validate(model, train_loader, val_loader, criterion, optimizer):
    """
    Function to run training and validation
    Parameters:
        model: The pytorch model object
        train_loader: PyTorch dataloader corresponding to the training dataset
        val_loader: PyTorch dataloader corresponding to the validation dataset
        criterion: The PyTorch loss function
        optimizer: The PyTorch optimizer to use for weight update
    Return:
        validation loss: The average of validation loss across validation set
    """
    model.train()
    epochs = 1
    # Training
    for i in range(epochs):
      for inputs, targets in train_loader:
          # Zero the gradients before forward pass
          optimizer.zero_grad()
          inputs = inputs.to("cuda")
          targets = targets.to("cuda")
          outputs = model(inputs)
          loss = criterion(outputs, targets)
          # Backpropagation
          loss.backward()
          optimizer.step()

    # Validation
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for inputs, targets in val_loader:
            inputs = inputs.to("cuda")
            targets = targets.to("cuda")
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            val_loss += loss.item()
    # Not returning the model as the weights as updated inside the model class
    return val_loss / len(val_loader)

# Main function for cross-validation
def cross_validate(trainset, valset, input_size, batch_sizes, hidden_layers_options, window_size):
    """
    Function to run cross validation
    Parameters:
        trainset: The training dataset object
        valset: The validation dataset object
        batch_size: List of all the batch_sizes for cross validation
        hidden_layers_options: List of hidden layer options for cross validation. Has both hidden size and num of layers
        window_size: Size of the context window
    Returns:
        loss_results: Dictionary containing the parameters tested and the respective 5-fold validation loss
    """
    # 5 fold cross validation
    kfold = KFold(n_splits=5, shuffle = True, random_state = 42)
    #Storing results
    loss_results = {}
    window_size = window_size
    validation_set = torch.utils.data.DataLoader(valset, batch_size = 1, shuffle = False, num_workers = 0)
    # Iterating through all batch sizes and hidden layer configurations
    for batch_size in batch_sizes:
        for hidden_layers in hidden_layers_options:

            # List for keeping track of loss per fold
            fold_loss = []
            # Splitting datasaet based on Kfold
            for train_idx, val_idx in kfold.split(trainset):
                train_subset = torch.utils.data.Subset(trainset, train_idx)
                val_subset = torch.utils.data.Subset(trainset, val_idx)
                # Dataloader Creation with new train and val
                # Batch size is utilized here
                train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers = 0)
                val_loader = DataLoader(val_subset, batch_size=1, shuffle = False, num_workers = 0)
                # Initializing model
                # Hidden layer options used here
                model = LSTMModel(input_size=input_size, hidden_size = hidden_layers[0], num_layers = hidden_layers[1])
                model.to("cuda")
                # Model training parameters
                criterion = nn.MSELoss()
                optimizer = optim.Adam(model.parameters(), lr=0.001)
                # Obtaining validation loss for this fold
                val_loss = train_and_validate(model, train_loader, val_loader, criterion, optimizer)
                fold_loss.append(val_loss)
            # Averaging loss across the folds
            avg_loss = sum(fold_loss) / len(fold_loss)
            string = "Batch Size: " + str(batch_size) + "; LSTM: {} hidden size and {} layers".format(hidden_layers[0], hidden_layers[1])
            loss_results[string] = avg_loss

    return loss_results
```

## Prediction Functions

The given prediction function predicts t time steps ahead. Since the input is a time series data, the 0th element in the series is removed and the newly predicted difference is appended to the time series and adjusted accordingly for predicting the next step.

```python
# function for running a advanced prediction across t steps
def t_time_step(tensor, model, inmf, t):
    """
    Function for predicting t time steps ahead using the model
    Parameters:
        tensor: The starting tensor
        model: The PyTorch model
        inmf: The input membership function -- required for fuzzification of the output difference
        t: time steps in the future we need to predict
    Returns:
        cumulative_difference: Tensor with the cumulative difference value across t time steps
    """
    # Variable to keep track of predicted difference
    cumulative_diff = 0
    # Going through each time step
    for i in range(t):
        # Output from the model for price difference time series
        outputs = model(tensor)
        diff = outputs.cpu().detach().numpy()
        # Adding the predicted difference to the cumulative difference
        cumulative_diff += diff
        # For next time step, removing difference in 0th position and adding predicted diff
        diff = fuzzification(diff, inmf, len(inmf[0]))
        diff_t = torch.tensor(diff).float().permute(1, 0)
        diff_t = diff_t.unsqueeze(0).to("cuda")
        tensor = tensor[:, 1:, :]
        tensor = torch.cat((tensor, diff_t), dim = 1)
    return cumulative_diff
```

The below function will use the trained model to predict the stock price t time steps ahead. The goal is to use the predicted price to calculate slow and fast moving average.

```python
def advanced_predictions(model, stock_price, inmf, dataloader, t, window_size):
    """
    Function to predict the price t time steps into the future
    Parameters:
        model: PyTorch model object
        stock_price: List containing the actual stock prices
        inmf: Input membership function
        dataloader: Data loader containing the stock price differences
        t: Time steps for predicting the future stock price
        window_size: Size of the context window
    Returns:
        pred: List containing the stock price predictions t time steps ahead
    """
    counter1 = window_size - 1
    pred = []
    for i, data in enumerate(dataloader, 0):
        inputs, labels = data
        inputs = inputs.to("cuda")
        labels = labels.to("cuda")
        # Finding cumulative difference across t time steps
        cumulative_diff = t_time_step(inputs, model, inmf, t)
        t_step_prediction = stock_price[counter1] + cumulative_diff
        counter1 += 1
        pred.append(t_step_prediction[0][0])
    return pred
```

## Returns Calculation Function

```python
def returns(df, column):
    """
    Function for calculating returns
    Parmeters:
        df: Dataframe containing the stock price information
        column: Column name used to retrieve FMA and SMA
    Returns:
        df: Dataframe edited with new columns with regards to returns
    """
  # Calculating Moving Averages based on predicted price
  df["Fast_SMA_" + column] = df[column].rolling(window = 20).mean()
  df["Slow_SMA_" + column] = df[column].rolling(window = 100).mean()
  # Signals -- Buy if Fast SMA > Slow SMA
  df["Signal_" + column] = 0
  df["Signal_" + column] = np.where(df['Fast_SMA_' + column] > df['Slow_SMA_' + column], 1.0, 0.0)
  # Calculate Daily Returns
  df["Daily_Return_" + column] = df[column].pct_change()
  # drop NA values
  df.dropna(inplace = True)
  # Calculate effective returns
  df["Effective_Return_" + column] = df["Daily_Return_" + column] * df["Signal_" + column]
  # Calculate cumulative return
  df['Cumulative_Return_' + column] = (1 + df['Effective_Return_' + column]).cumprod() - 1
  return df
```
